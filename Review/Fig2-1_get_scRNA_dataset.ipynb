```python
# GSE180286

import pandas as pd
import gzip
import os

file_name = "GSM5457213_E2020-3.expression_matrix.txt.gz"
file_path = os.path.join("./GSE180286_RAW/", file_name)

SC_df = pd.read_csv(file_path, sep="\t", compression="gzip")

SC_df
```

```python
SC_df_T = SC_df.T
SC_df_T
```

```python
import anndata
import pandas as pd
import gzip
import os

filepath = '../dataset/Dataset32/Insitu_count.txt'

ST_df = pd.read_csv(filepath, sep='\t', encoding='utf-8')

ST_df
```

```python
df_columns = set(SC_df_T.columns)
df_new_columns = set(ST_df.columns)

common_columns = df_columns.intersection(df_new_columns)
print(len(common_columns))

dataframe2_intersected = SC_df_T[list(common_columns)]

dataframe2_intersected
```

```python
import scanpy as sc

adata = sc.AnnData(dataframe2_intersected)

def leiden_clustering(adata, resolution=0.5, seed=1234):

    sc.tl.pca(adata, random_state=seed)
    
    sc.pp.neighbors(adata, random_state=seed)
    
    sc.tl.leiden(adata, random_state=seed, resolution=resolution)
    return adata

adata = leiden_clustering(adata, resolution=0.5, seed=1234)

adata
```

```python
adata.write('./dataset/Dataset32/Review_scRNA_count_cluster_new0.h5ad')
```

# GSE167036

```python
import pandas as pd
import numpy as np
from scipy.io import mmread

data_dir = "./GSE167036"
matrix_file = f"{data_dir}/GSE167036_cell_counts_matrix.mtx"
barcodes_file = f"{data_dir}/GSE167036_barcodes.csv.gz"
features_file = f"{data_dir}/GSE167036_features.csv.gz"

matrix = mmread(matrix_file).T

print(f"Matrix shape: {matrix.shape}")

barcodes = pd.read_csv(barcodes_file, compression='gzip', header=None, names=["barcodes"])
features = pd.read_csv(features_file, compression='gzip', header=None, names=["gene_symbols"])

print(f"Barcodes shape: {barcodes.shape}")
print(f"Features shape: {features.shape}")

if barcodes.shape[0] > matrix.shape[0]:
    print("Removing extra row from barcodes file.")
    barcodes = barcodes.iloc[1:]

if features.shape[0] > matrix.shape[1]:
    print("Removing extra row from features file.")
    features = features.iloc[1:]

print(f"Barcodes shape after cleaning: {barcodes.shape}")
print(f"Features shape after cleaning: {features.shape}")

if barcodes.shape[0] != matrix.shape[0] or features.shape[0] != matrix.shape[1]:
    raise ValueError("条形码或基因信息的行数与矩阵的行数或列数不匹配！")

df = pd.DataFrame(matrix.toarray(), index=barcodes["barcodes"], columns=features["gene_symbols"])

print("\nGenerated DataFrame:")

df
```

```python
import pandas as pd

df_columns = set(df.columns)

df_new_columns = set(ST_df.columns)

common_columns = df_columns.intersection(df_new_columns)
print(len(common_columns))
dataframe2_intersected = df[list(common_columns)]

dataframe2_intersected
```

```python
import scanpy as sc

adata = sc.AnnData(dataframe2_intersected)

def leiden_clustering(adata, resolution=0.5, seed=1234):

    sc.tl.pca(adata, random_state=seed)
    
    sc.pp.neighbors(adata, random_state=seed)
    
    sc.tl.leiden(adata, random_state=seed, resolution=resolution)
    return adata

adata = leiden_clustering(adata, resolution=0.5, seed=1234)

adata
```
