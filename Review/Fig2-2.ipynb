```python
import matplotlib.pyplot as plt

K = [5, 10, 15, 20, 25, 30]
RMSE = [0.784, 0.762, 0.746, 0.749, 0.753, 0.737]
JS = [0.118, 0.114, 0.117, 0.110, 0.119, 0.113]

baseline_RMSE = 0.818
baseline_JS = 0.139

fig, ax1 = plt.subplots(figsize=(9, 6))


ax1.plot(K, RMSE, marker='o', label='SpaIM (RMSE)', color='#1f77b4', markersize=8)
ax1.set_xlabel('K', fontsize=16)
ax1.set_ylabel('RMSE Score', fontsize=16, color='#1f77b4') 
ax1.tick_params(axis='both', labelsize=14) 
ax1.tick_params(axis='y', labelcolor='#1f77b4')
ax1.grid(True)

ax2 = ax1.twinx()

ax2.plot(K, JS, marker='s', label='SpaIM (JS)', color='#2ca02c', markersize=8)
ax2.set_ylabel('JS Score', fontsize=16, color='#2ca02c') 
ax2.tick_params(axis='y', labelcolor='#2ca02c', labelsize=14) 

ax1.axhline(y=baseline_RMSE, color='#17becf', linestyle='--', label='Baseline (Tangram)')
ax2.axhline(y=baseline_JS, color='#17becf', linestyle='--') 

plt.title('Breast Cancer', fontsize=18, pad=20)

lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()

all_lines = lines + lines2
all_labels = labels + labels2
fig.legend(all_lines, all_labels, loc='upper center', bbox_to_anchor=(0.7, 0.84), ncol=1, fontsize=14)

plt.tight_layout()

plt.savefig('FigK_32_Breast Cancer.pdf', format='pdf', bbox_inches='tight', dpi=300)

plt.show()
```
