```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams.update({'font.size': 16})

datasets = ['Default', 'GSE180286', 'GSE167036', 'Default', 'GSE180286', 'GSE167036']
SSIM_values = [0.612, 0.603, 0.614]
js_values = [0.118, 0.105, 0.122]


SSIM_color = '#A2D9CE'
js_color = '#D2B48C'


x = np.arange(len(datasets))


fig, ax1 = plt.subplots(figsize=(10, 6))


rects1 = ax1.bar(x[:3], SSIM_values, width=0.5, label='SSIM', color=SSIM_color)


ax1.set_ylim(0, 0.8)
ax1.set_ylabel('SSIM Scores', color=SSIM_color, fontsize=18)
ax1.tick_params(axis='y', labelcolor=SSIM_color, labelsize=16)

ax2 = ax1.twinx()

rects2 = ax2.bar(x[3:], js_values, width=0.5, label='JS', color=js_color)

ax2.set_ylim(0, 0.3)
ax2.set_ylabel('JS Scores', color=js_color, fontsize=18) 
ax2.tick_params(axis='y', labelcolor=js_color, labelsize=16)

ax1.set_xticks(x)
ax1.set_xticklabels(datasets, ha='center', fontsize=14)

ax1.set_xlabel('Datasets', fontsize=18) 
ax1.set_title('Breast Cancer', fontsize=20)

from matplotlib.patches import Patch
legend_elements = [
    Patch(facecolor=SSIM_color, label='SSIM'),
    Patch(facecolor=js_color, label='JS')
]
ax1.legend(handles=legend_elements, fontsize=16) 

def autolabel(rects, ax):
    for rect in rects:
        height = rect.get_height()
        ax.annotate(f'{height:.3f}',
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom', fontsize=14) 

autolabel(rects1, ax1)
autolabel(rects2, ax2)

fig.tight_layout()

plt.savefig('SFig7b_CID44971.pdf', format='pdf', bbox_inches='tight', dpi=300)

plt.show()
```
