### CID44971 JS

```python
import matplotlib.pyplot as plt

 
K = [5, 10, 15, 20, 25, 30]
JS = [0.118, 0.114, 0.117, 0.110, 0.119, 0.113]
baseline = 0.124

  
plt.rcParams.update({'font.size': 16})

   
plt.figure(figsize=(6, 4))

    
plt.plot(K, JS, 
         marker='o', 
         markersize=8,
         linewidth=2.5,
         color='#C16A5C',   
         label='SpaIM')

 
plt.axhline(y=baseline, 
            color='#CEA26B', 
            linestyle='--', 
            linewidth=2,
            label='Baseline')

 
plt.ylim(0.1, 0.14)

 
plt.xlabel('K Value', fontsize=12, labelpad=10)
plt.ylabel('JS', fontsize=12, labelpad=10)
plt.xticks(K, fontsize=11)
plt.yticks(fontsize=11)

 
plt.grid(True, linestyle='--', alpha=0.4)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

  
plt.legend(fontsize=12, framealpha=0.9)

  
plt.tight_layout()
plt.savefig('CID44971_JS.pdf', format='pdf', bbox_inches='tight', dpi=300)

plt.show()
```
